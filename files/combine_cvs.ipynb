```python
import os
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
data_arquivo_folder = 'C:/Users/Utilizador/Desktop/Bianca/CURSOS EXTRAS/CIENTISTA DE DADOS - DNC/PROJETOS/Projeto Casting/DADOS ORIGINAIS/data_explorer/data'

df = []

for file in os.listdir(data_arquivo_folder):
    if file.endswith('.csv'):
        print('Loading file {0}...'.format(file))
        df.append(pd.read_csv(os.path.join(data_arquivo_folder, file))) 

print(len(df)) 

```

```python
df_principal = pd.concat(df, axis=0)

df_principal.to_csv('C:/Users/Utilizador/Desktop/Bianca/CURSOS EXTRAS/CIENTISTA DE DADOS - DNC/PROJETOS/Projeto Casting/DADOS ORIGINAIS/data_explorer/data/casting_total.csv', index=False)
```

```python
# AJUSTAR TIPO COLUNA: VUF_DT 

df_principal.VUF_DT = pd.to_datetime(df_principal.VUF_DT)
```

```python
# DROPAR COLUNAS QUE NÃO SERÃO UTILIZADAS NO MODELO:

df_principal = df_principal.drop(['VUF_CODIGO_BOLETO', 'VUF_QTBOLETO', "VUF_VLRTROCA", "VUF_VLRDESCONTO"], axis=1)
```

```python
# FILTRAR VALORES NEGATIVOS:

df_principal = df_principal.query('VUF_QTPRODUTO > 0' and 'VUF_VLRBRUTOVENDA > 0' and 'VUF_VLRLIQFINAL > 0') 
```

```python
# REMOVER DADOS NULOS

df_principal = df_principal.dropna() 
```

```python
# agrupar valor de faturamento sem desconto por dia:

df_fat_diario = df_principal.groupby("VUF_DT")["VUF_VLRBRUTOVENDA"].sum()
df_fat_diario.head()
```

```python
# Faturamento ao longo do tempo

plt.figure( figsize=(23,6) )
plt.plot( df_fat_diario)
plt.title('Faturamento ao longo do tempo')
plt.grid(linestyle='dotted')

plt.xlabel('Data')
plt.ylabel('Faturamento');
```

```python
# VENDAS POR MêS:

vendas_mês = df_principal.groupby(df_principal["VUF_DT"].dt.month)['VUF_VLRLIQFINAL'].sum().sort_values(ascending=False).plot(kind="bar", color="green", figsize=(20,5), xlabel="MÊs da Venda", ylabel="Valor Vendas (R$)")
plt.title("Vendas por MÊs", fontsize=14, fontweight="bold")
plt.xticks(rotation=45)
plt.grid(linestyle='dotted')

plt.show()
```

```python
# VENDAS POR loja:

VENDAS_LOJAS = df_principal.groupby(df_principal["UND_CODIGO"])['VUF_VLRLIQFINAL'].sum().nlargest(20).plot(kind="bar", color="green", figsize=(20,5), xlabel="Código da Loja", ylabel="Valor Vendas (R$)")
plt.title("20 Lojas com maiores vendas", fontsize=14, fontweight="bold")
plt.xticks(rotation=75)
plt.grid(linestyle='dotted')

plt.show()
```

```python
# VENDAS POR CLIENTE:

VENDAS_CLIENTE = df_principal.groupby("CLV_BANCO")['VUF_VLRLIQFINAL'].sum().sort_values(ascending=False).plot(kind="bar", color="green", figsize=(20,5), xlabel="Cliente", ylabel="Valor Vendas (R$)")
plt.title("Faturamento por cliente", fontsize=14, fontweight="bold")
plt.xticks(rotation=75)
plt.grid(linestyle='dotted')
plt.show()
```

```python
# VENDAS POR FUNCIONARIO:

VENDAS_FUNCIONARIO = df_principal.groupby("FUN_CODIGO")['VUF_VLRLIQFINAL'].sum().nlargest(20).plot(kind="bar", color="green", figsize=(20,5), xlabel="Funcionario", ylabel="Valor Vendas (R$)")
plt.title("20 Funcionários com maiores vendas", fontsize=14, fontweight="bold")
plt.xticks(rotation=75)
plt.grid(linestyle='dotted')

plt.show()
```
